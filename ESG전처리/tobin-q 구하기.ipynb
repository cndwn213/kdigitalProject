```python
jj_df = pd.read_csv("data/재무제표2.csv")
# jj_df["고유번호"] = jj_df["고유번호"].astype("str")
jj_df.loc[7803:7885,"연도"] = 2012
jj_df.loc[8075:8079,"연도"] = 2012
display(jj_df[jj_df["연도"].isna()])

jj_df= jj_df.astype({"연도":"int"})
jj_df.info()
jj_df.drop(jj_df[jj_df["연도"]==2010].index, inplace=True)
jj_df[jj_df["연도"]==2010]
jj_df.reset_index(drop=True, inplace = True)
jj_df.tail(5)
```

```python
jj_df["기업명"].duplicated().sum()
```

```python
#토빈Q
tobin_lst = []
for i in jj_df.index:
    value = sc_df[sc_df["종목명"] == jj_df.iloc[i,0]][str(jj_df.iloc[i,-1])].values
    if list(value):
        tobin = value[0]/jj_df.iloc[i,4] * 100
        tobin_lst.append([jj_df.iloc[i,0], sc_df[sc_df["종목명"] == jj_df.iloc[i,0]]["종목코드"].values[0], tobin, jj_df.iloc[i,-1]])
    else:
        tobin_lst.append([jj_df.iloc[i,0], 0,0,jj_df.iloc[i,-1]])
tobin_lst
```

```python
tobin_df = pd.DataFrame(tobin_lst, columns = ["종목명","종목코드","tobinq","연도"])
index = tobin_df["종목명"].unique()
tobin_df.fillna(0, inplace = True)
len(index)
```

```python
code_lst = []
for name in index:
    code_lst.append(tobin_df[tobin_df["종목명"] == name]["종목코드"].values[0])
len(code_lst)
```

```python
index_df = pd.DataFrame({"종목명":index, "종목코드":code_lst})
index_df.head(5)
```

```python
for i in range(2011, 2022):
    tobin_lst = []
    for name in index:
        tobin = tobin_df[(tobin_df["종목명"] == name) & (tobin_df["연도"] == i)]["tobinq"].values
        if list(tobin):
            tobin_lst.append(tobin[0])
        else: tobin_lst.append(0)    
    print(len(tobin_lst))
    index_df = pd.concat([index_df, pd.DataFrame(tobin_lst)],axis= 1)
```

```python
index_df.columns = ["종목명","종목코드","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"]
```

```python
lst = []
for i in df["기업명"].unique():
    val = index_df[index_df["종목명"] == i]["종목명"].values
    if list(val):
        lst.append(index_df[index_df["종목명"] == i]["종목명"].values[0])
len(lst)
```
