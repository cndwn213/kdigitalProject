```python
import pandas as pd
import requests 
import numpy as np
from bs4 import BeautifulSoup as bs
import time
```

## KCG홈페이지 자료(2021)

```python

dt_lst = []
for i in range(1, 102):
    req = requests.get("http://www.cgs.or.kr/business/esg_tab04.jsp?pg="+str(i)+"&pp=10&skey=&svalue=&sfyear=2021&styear=2021&sgtype=&sgrade=#ui_contents")
    soup = bs(req.content, "html.parser")
    table = soup.find("div", class_ ="business_board")
    tr_lst = table.find_all("tr")
    tr_lst = tr_lst[1:]
    for tr in tr_lst :
        td_lst = tr.find_all("td")
        text_lst = []
        for td in td_lst:
            text_lst.append(td.get_text())
        dt_lst.append(text_lst[:-2])
        
df = pd.DataFrame(dt_lst, columns = ["NO","기업명","기업코드","ESG등급","환경","사회","지배구조"])

# df.sort_values(by="기업코드",ascending = True) # 기업 코드로 정렬
df.sort_values(by="기업명",ascending = True)     # 기업명으로 정렬

# df.to_csv("esg기업 목록") # csv파일 내보내기
```
