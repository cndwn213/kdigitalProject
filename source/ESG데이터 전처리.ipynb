```python
# 모든 년도에 걸쳐 존재하는 기업 이름 찾기
all_lst = []
for j in range(8):
    
    lst = np.zeros(len(name_code), dtype= np.int64)
    for i in range(len(name_code["기업명"])):
        name = name_code.iloc[i,0]
        for df_ in df_lst[j:]:
            if name not in df_["기업명"].unique():
                lst[i] = 1
    in_df = pd.DataFrame(lst)
    in_ = in_df[in_df[0] != 1]
    all_lst.append(in_)
    print(str(j+11)+"년부터 연속", len(in_))
```

```python
name_all =[]
for v in all_lst:
    index = v.index
    name_ = [name_code.iloc[i,0] for i in index]
    name_all.append(name_)
    print(len(name_))
```

```python
for i, v in enumerate(name_all):
    df_ = pd.DataFrame(v)
    df_.to_csv("data/from_"+str(i + 11)+".csv", index= False)
```
